## **Data visualization using Matplotlib and Seaborn**

##### Use  `matplotlib.pyplot` for basic plots like line and bar charts. It is imported with the alias `plt`. The `seaborn` module will be used for more advanced plots and it is imported with the alias `sns`.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### **Plotting Multiple Lines**

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.939, 0.941, 0.942]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.90, 0.901, 0.901]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

### **Chart Title and Legend**

##### To differentiate between multiple lines, we can include a legend within the graph using `plt.legend` function. We can also give the entire chart a title using `plt.title` function.

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');

plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges']);
```

### **Line Markers**

##### Use markers for data points on each line using the `marker` argument of `plt.plot`. Matplotlib supports many different types of makers like circle, cross, square, diamond, etc.

```python
plt.plot(years, apples, marker = 'o')
plt.plot(years, oranges, marker = 'x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');

plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges']);
```

##### Refer this link for different marker types:  https://matplotlib.org/3.1.1/api/markers_api.html

### **Styling lines and markers**

##### The`plt.plot` function supports many arguments for styling lines and markers:

- `color` or `c` : set the color of the line [(Supported colors)](https://matplotlib.org/3.1.0/gallery/color/named_colors.html)
- `linestyle` or `ls` : choose between a solid or dashed line
- `linewidth` or `lw` : set the width of a line
- `markersize` or `ms` : set the size of markers
- `markeredgecolor` or `mec` : set the edge color for markers
- `markeredgewidth` or `mew` : set the edge width for markers
- `markerfacecolor` or `mfc` : set the fill color of markers
- `alpha` : opacity of the plot

Check the below link for `plt.plot` to lern more:
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

```python
plt.plot(years, apples, marker = 's', c = 'b', ls = '-', lw = 2, ms = 8, mew = 2, mec = 'navy')
plt.plot(years, oranges, marker = 'o', c = 'r', ls = '--', lw = 3, ms = 10, alpha = .5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');

plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges']);
```

##### `fmt` argument provides a shorthand for specifying the line style, marker and line color. It can be provided as the 3rd argument to `plt.plot`

                fmt = '[marker][line][color]'

```python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');

plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges']);
```

##### If no line style is specified in `fmt`, only markers are drawn

```python
plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare)");
```

### **Changing the figure size**

##### Use `plt.figure` function to change the size of the figure.

```python
plt.figure(figsize = (8,4))

plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare)");
```

## **Scatter Plot**

##### A scatter plot is used to visualize the relationship between two variables as points on a 2-dimensional grid.

```python
import pandas as pd
```

```python
flowers_df = sns.load_dataset('iris')
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

##### Lets's try to visualize the relationship between sepal length and sepal width by creating a line chart using `plt.plot`.

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
```

##### As the above plot is not so informative, we can use the scatter plot to visualize the relationship using the `scatterplot` function from `seaborn` (imported as `sns`).

```python
#sns.scatterplot(flowers_df.sepal_length, flowers_df.sepal_width);
```

## **Histogram**

##### Use `plt.hist` function to create a histogram. Histograms's are specifically used to visualize the distribution for a specific column.

```python
flowers_df.sepal_width
```

```python
flowers_df.describe()
```

```python
plt.title('Distribution of Sepal Width')
plt.hist(flowers_df.sepal_width);
```

##### It is splitted in 10 intervals (i.e 10 bins).

### **Controlling the size and no. of bins**

##### We can control the no. of bins, or the size of each bin using the `bins` argument.

```python
#specifying the no. of bins
plt.hist(flowers_df.sepal_width, bins = 5);
```

```python
import numpy as np

#specifying the boundries of each bin
plt.hist(flowers_df.sepal_width, bins = np.arange(2, 5, 0.25));  #start:stop(exclsuive):step
```

```python
#bins of unequal sizes
plt.hist(flowers_df.sepal_width, bins = [1, 3, 4, 4.5]);
```

### **Multiple Histograms**

##### Similar to line charts we can draw multiple histograms in a single chart. We can reduce the opacity of each histogram, so that the bars of one histogram don't hide for others.

##### Separate histograms for each species of flowers.

```python
setosa = flowers_df[flowers_df.species == 'setosa']
versicolor = flowers_df[flowers_df.species == 'versicolor']
virginica = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa.sepal_width, alpha = 0.4, bins = np.arange(2, 5, 0.25));
plt.hist(versicolor.sepal_width, alpha = 0.4, bins = np.arange(2, 5, 0.25));

plt.legend(['Setosa','Versicolor'])
```

##### We can also stack multiple histograms on top of one another using the `stacked` argument in the `plt.hist` function.

```python
plt.title('Distribution of Sepal Width')

plt.hist([setosa.sepal_width, versicolor.sepal_width, virginica.sepal_width],
        bins = np.arange(2, 5, 0.25),
        stacked = True);

plt.legend(['Setosa','Versicolor','Virginica']);
```

## **Bar Chart**

## **Heatmap**

##### A heatmap is used to visualize a 2-dimensional data like matrix or a table using colors. 

```python
df_flights = sns.load_dataset('flights')
```

```python
plt.plot(df_flights.passengers)
```

```python
#df_flights_pivot = sns.load_dataset('flights').pivot('month','year','passengers')
```

```python
df_flights_pivot = df_flights.pivot(index="month", columns="year", values="passengers")
df_flights_pivot
```

##### We can use the `sns.heatmap` function to visualize the above data.

```python
plt.title('No. of Passangers (1000s)')
sns.heatmap(df_flights_pivot);
```

##### We can also display the actual values in each block by specifying `annot = True`, and use the `cmap` argument to change the color palette.

```python
plt.title('No. of Passangers (1000s)')
sns.heatmap(df_flights_pivot, fmt = 'd', annot = True, cmap = 'Blues');
```
